```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load
import pandas as pd

input_data = pd.read_csv('mushrooms.csv')
```

Now split into train an test

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(input_data, test_size=0.2)
```

Train shape

```python
train_df.shape
```

Test shape

```python
test_df.shape
```

Selecting and dropping target column for train dataset

```python
train_target = train_df['class']
train_features = train_df.drop('class', axis=1)
train_features
```

Selecting and dropping target column for test dataset

```python
test_target = test_df['class']
test_features = test_df.drop('class', axis=1)
test_features
```

One hot encoding

```python
test_features = pd.get_dummies(test_features)
train_features = pd.get_dummies(train_features)
```

Binary class encoding

```python
from sklearn.preprocessing import LabelEncoder

test_target = LabelEncoder().fit_transform(test_target)
train_target = LabelEncoder().fit_transform(train_target)
```

Try models

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)
rf_classifier.fit(train_features, train_target)

rf_prediction = rf_classifier.predict(test_features)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(train_features, train_target)
dt_prediction = dt_classifier.predict(test_features)
```

XGBoost

```python
from xgboost import XGBClassifier

xgb_classifier = XGBClassifier()
xgb_classifier.fit(train_features, train_target)
xgb_prediction = xgb_classifier.predict(test_features)
```

Check accuracy score

```python
from sklearn.metrics import accuracy_score

rf_accuracy = accuracy_score(rf_prediction, test_target)
print("Random Forest Classifier Accuracy:", rf_accuracy)

dt_accuracy = accuracy_score(dt_prediction, test_target)
print("Decision Tree Classifier Accuracy:", dt_accuracy)

xgb_accuracy = accuracy_score(xgb_prediction, test_target)
print("XGBoost Classifier Accuracy:", xgb_accuracy)
```

Model Explainability

```python
importance_df = pd.DataFrame({'Feature': train_features.columns, 'Importance': rf_classifier.feature_importances_})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
importance_df = pd.DataFrame({'Feature': train_features.columns, 'Importance': dt_classifier.feature_importances_})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
importance_df = pd.DataFrame({'Feature': train_features.columns, 'Importance': xgb_classifier.feature_importances_})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)
```
